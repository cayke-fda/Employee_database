```python
#For this project we will be using the Employee dataset dataset from Kaggle
#It is available at https://www.kaggle.com/datasets/tawfikelmetwally/employee-dataset

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Employee.csv')
df
```

```python
list_education = []
for elem in df['Education']:
    if elem not in list_education:
        list_education.append(elem)
list_education
```

```python
#We start by assigning values to candidates who have a PhD (PHD=1),just a Masters (PHD=0 and Masters = 1)
#or just a Bachelors (PHD=0 and Masters = 0)
df.loc[(df['Education'] == 'Masters') , 'Masters'] = 1
df.loc[(df['Education'] != 'Masters'), 'Masters'] = 0

df.loc[df['Education'] == 'PHD', 'PHD'] = 1 
df.loc[df['Education'] != 'PHD', 'PHD'] = 0

df.drop('Education',axis=1, inplace = True)

df
```

```python
#Next we assign a collumn that assigns a vaue if the candidate identifies with the female gender

df.loc[df['Gender'] == 'Female', 'Female'] = 1
df.loc[df['Gender'] == 'Male', 'Female'] = 0

df.drop('Gender', axis=1, inplace = True)
df
```

```python
#Finally we assign a collumn that assigns a vaue if the candidate has ever been benched

df.loc[df['EverBenched'] == 'Yes', 'Benched'] = 1
df.loc[df['EverBenched'] == 'No', 'Benched'] = 0

df.drop('EverBenched', axis=1, inplace = True)
df
```

```python
#Now we count the number of employees in each city

dict_city = {}
for city in df['City']:
    try:
        dict_city[city] += 1
    except:
        dict_city[city] = 1

display(dict_city)
#And now we do the same thing we did with their education level
df.loc[df['City'] == 'Bangalore', 'Bangalore'] = 1
df.loc[df['City'] != 'Bangalore', 'Bangalore'] = 0

df.loc[df['City'] == 'Pune', 'Pune'] = 1
df.loc[df['City'] != 'Pune', 'Pune'] = 0

df.drop('City', axis=1, inplace=True)
df
```

```python
sns.pairplot(df)
```

```python
#We plot a heatmap of the correlation between the different labels 
sns.heatmap(abs(df.corr()))
display(df.corr()['LeaveOrNot'])
```

```python
df.columns
```

```python
#It looks like that the features that matter the most to leave or stay are Pune, Female, 
#PaymentTier and JoiningYear

#We start our analyzes
from sklearn.model_selection import train_test_split
y = df['LeaveOrNot']
X = df[['JoiningYear', 'PaymentTier', 'Age', 'ExperienceInCurrentDomain',
        'Masters', 'PHD', 'Female', 'Benched', 'Bangalore','Pune']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression

logmodel = LogisticRegression(max_iter=10**3)
```

```python
logmodel.fit(X_train, y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
logmodel.coef_
```

```python
columns = list(df.columns.drop('LeaveOrNot'))
dict_coef = {}

for i in range(len(columns)):
    dict_coef[columns[i]] = logmodel.coef_[0][i]
dict_coef
```

```python
df.rename(columns={'ExperienceInCurrentDomain':'Experience'}, inplace = True)
columns = list(df.columns.drop('LeaveOrNot'))
plot = sns.barplot(x = columns, y = logmodel.coef_[0])
sns.set(rc={'figure.figsize':(11.7,8.27)})
plot
```

# Now we work on some questions

## What is the distribution of educational qualifications among employees?

```python
num_PhD = sum(df['PHD'])
num_Masters = sum(df['Masters'])
num_Bachelors = len(df['PHD']) - (num_PhD + num_Masters)
data = [num_Bachelors, num_Masters, num_PhD]
labels = ['Bachelors', 'Masters', 'PhD']

plt.pie(data, labels = labels, autopct='%.0f%%')
plt.title('Distribution of educational qualifications')
```

## How does the length of service (Joining Year) vary across different cities?

```python
JY_Bang, JY_Pune, JY_ND = [],[],[]
names = ['Bangalore', 'Pune', 'New Delhi']
for i in range (len(df)):
    if df['Bangalore'][i]==1:
        JY_Bang.append(df['JoiningYear'][i])
    elif df['Pune'][i]==1:
        JY_Pune.append(df['JoiningYear'][i])
    else:
        JY_ND.append(df['JoiningYear'][i])
JY_Bang, JY_Pune, JY_ND = np.array(JY_Bang),np.array(JY_Pune),np.array(JY_ND)
```
